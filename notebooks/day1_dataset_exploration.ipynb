Imports

```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
print("imports pass")
```

Define Paths

```python
dataset_path = Path("../data/raw/vep1/TESELLATED_WITHOUT_AUGMENTATION")

rgb_path = dataset_path / "RGB"
mask_path = dataset_path / "MASK"

print(rgb_path)
print(mask_path)
```

Check Number of Images

```python
rgb_images = sorted(list(rgb_path.glob("*")))
mask_images = sorted(list(mask_path.glob("*")))

print("RGB Images:", len(rgb_images))
print("Mask Images:", len(mask_images))
```

Load First Image

```python
image = cv2.imread(str(rgb_images[0]))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

print(image.shape)

plt.figure(figsize=(8,8))
plt.imshow(image)
plt.title("Sample RGB Image")
plt.axis("off")
plt.show()
```

Load First Mask

```python
mask = cv2.imread(str(mask_images[0]), cv2.IMREAD_GRAYSCALE)

print(mask.shape)

plt.figure(figsize=(8,8))
plt.imshow(mask, cmap="gray")
plt.title("Sample Mask")
plt.axis("off")
plt.show()
```

Check Mask Values

```python
print("Unique mask values:", np.unique(mask))
```
